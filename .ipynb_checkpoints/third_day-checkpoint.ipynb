**1：打印一个矩形，用什么符号表示矩形自行定义, 矩形的宽度值需要用户自定义**

```python
a = input("请输入矩形的宽度：")
for i in range(int(a)):
    if i == 0 or i == 19:
        print("*"*20)
        
    else:
        print("*"+" "*18+"*")
```

2：计算出1到n阶层之和，使用input函数

```python
# 第一种做法，两层循环实现
a = int(input("请输入一个数"))
sum = 0
for i in range(1,a+1):
    cur = 1
    for j in range(1,i+1):
        cur = cur * j
    sum = sum + cur
print(sum)
```

```python
# 第二种做法，一层循环实现
a = int(input("请输入一个数"))
sum = 0 
cur = 1
for i in range(1,a+1):
    cur = cur * i
    sum = sum + cur
print(sum)
```

3：输入一个数，判断是否是质数（质数定义为在大于1的自然数中，除了1和它本身以外不再有其他因数，1不是质数也不是和数）

```python
a = int(input("请输入一个数："))

if a == 2:
    print("这个数是质数")
else:
    for i in range(3, a):
        if a % i == 0:
            print("%d不是质数" % a)
            break
    else:
        print("%d是质数" % a)

            
```

```python
import time
start = time.time()
for a in range(2,10000):
    for i in range(2, a):
        if a % i == 0:
            print("%d不是质数" % a)
            break
    else:
        print("%d是质数" % a)
end = time.time()
print(end - start)
```

```python

n = int(input("请输入一个数字："))
a = 0  
b = 1  
for i in range(1,n): 
    if i == 1:
        print(1, end=" ")
    c = a+b  
    a, b = b, c
    #a = b  
    #b = c   
    print(c, end=" ")  
```

```python
# 介绍第一个模块，time模块

import time
a = 0
b = 1
start_time = time.time()
print("开始时间为：%s" % start_time)
for i  in range(1, 4):
    if i == 1: 
        print(i, end = " ")
    c = a + b
    a = b
    b = c
    print(c, end = " ")
end_time=time.time()
print("\n结束时间为：%s" % end_time)
print("消耗时间为：%f" % (end_time - start_time))
```

7: 说一说print格式化

```python
print('{0} 讨厌 {1}'.format("傻逼", "牛逼"))
```

```python
print('%s 喜欢 %s' % ("牛逼", "傻逼"))
```

```python

for i in range(9, 0, -1):
    for j in range(1,i+1):
        # 这里的{2:<6}中，6表示j*i占6个位置，而默认是右对齐的，因此让6个位置左对齐
        print("{0}X{1}={2:<6}".format(j, i, j*i), end = "")
    print()
```

```python
import time
start = time.time()
print(1)
print(2)
print(3)
for i in range(3, 10000, 2):
    m = i ** 0.5 
    for j in range(3, i , 2):
        if i % j == 0:
            break
        if j > m:
            print(i)
            break
end = time.time()
print(end - start)
```
